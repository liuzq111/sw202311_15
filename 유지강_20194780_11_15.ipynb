라이브러리 불러오기

```python
import cv2
from google.colab.patches import cv2_imshow
```

# 이미지 불러오기

```python
img=cv2.imread('/content/drive/MyDrive/11_15_이미지.jpg')
cv2_imshow(img)
```

# 이미지를 그레이스케일로 변환

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

# Canny 방법 - 원본 이미지 그대로 가장자리를 추출하는 방법

```python
method_canny = cv2.Canny(img, 100, 255)
```

# Canny로 방법 으로 가장자리 추출한 이미지

```python
cv2_imshow(method_canny)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
