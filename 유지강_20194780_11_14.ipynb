```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
img=cv2.imread('/content/drive/MyDrive/lzqMerchData/Key Board/Key1.jpg')
cv2_imshow(img)
img_up=cv2.resize(img,None, fx=2.0, fy=2.0, interpolation=cv2.INTER_LINEAR)
cv2_imshow(img_up)
img_down=cv2.resize(img,None, fx=0.5, fy=0.5, interpolation=cv2.INTER_LINEAR)
cv2_imshow(img_down)
img_croped=img[:,:128]
cv2_imshow(img_croped)
(b, g, r) = img[0, 0]
print("Pixel at (0, 0) - Red: {}, Green: {}, Blue: {}".format(r, g, b))
cv2.imwrite('/content/drive/MyDrive/lzqMerchData/Key Board/img_up.jpg',img_up)
cv2.imwrite('/content/drive/MyDrive/lzqMerchData/Key Board/img_down.jpg',img_down)
cv2.imwrite('/content/drive/MyDrive/lzqMerchData/Key Board/img_croped.jpg',img_croped)
```
